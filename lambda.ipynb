## Lambda Functions


## 1

```python
#Using Lambda Functions
square = lambda x: x**2
print(square(5)) 
```

```python
#Traditional way
def square(x):
    return x**2
square(5)
```

## 2.


```python
#Using Lambda Functions
add = lambda a,b : a+b
print(add(2,5))
```

```python
##Traditonal way
def add(a,b):
    return a+b
print(add(2,5))
```

## 3.

```python
## using lambda functions
employees = [
    {"Name": "Navaneeth", "Age": 23},
    {"Name": "Sricharan", "Age": 22},
    {"Name": "Jayanth", "Age": 24}
]

sort_by_age = sorted(employees, key = lambda emp : emp["Age"])
print(sort_by_age)
```

```python
## Traditional Way
employees = [
    {"Name": "Navaneeth", "Age": 23},
    {"Name": "Sricharan", "Age": 22},
    {"Name": "Jayanth", "Age": 24}
]
n = len(employees) 
for i in range(n):
    for j in range(0, n - i - 1):
        if employees[j]["Age"] > employees[j + 1]["Age"]:
            temp = employees[j]             
            employees[j] = employees[j + 1] 
            employees[j + 1] = temp        
print(employees)
```
